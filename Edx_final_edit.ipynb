## About the data:

### The data has collected from ‘HarvardX’ and ‘MITx’ courses in ‘Edx’ website 
### which provide courses for students.
#### Column’s name Description
1. Institution: Names of online course holders.
2. Course Number: the unique id of each course.
3. Launch Date: the launch date of each course(year/month/day).
4. Course Title: the title of each course provided by institution.
5. Instructors: the instructors of each course.
6. Course Subject: the subject of each course.
7. Year: the last time of each course
8. Honor Code: Certificates with (1), without (0).
9. Participants (Course Content Accessed): the number of participants who have accessed the course.
10. Audited: (> 50% Course Content Accessed): the number of participants who have audited more than 50% of the course.
11. Certified: the number of participants who have been certified from the course.
12. % Audited: the percent of the audited %
13. % Certified: the percent of the certified %
14. % Certified of > 50% Course Content Accessed: the percent of the certified with accessing the course more than 50%
15. % Played Video: the percent of playing video of course.
16. % Posted in Forum: the percent of posting in forum.
17. % Grade Higher Than Zero: the percent of grade higher than zero.
18. Total Course Hours (Thousands): total course hours (per 1000).
19. Median Hours for Certification: median hours for certification from each 
course.
20. Median Age: median age of the participants in course.
21. % Male: the percent of the males participated in 
course.
22. % Female: the percent of the females participated in 
course.
23. % Bachelor's Degree or Higher: the percent of bachelor's degree of higher%

## Data Cleaning:

### extract the month and year from 'launch Date' column

```python
df['DateTime'] = pd.to_datetime(df['Launch Date'])
df['month'] = df['DateTime'].dt.month
df['month']=df['month'].astype('object')
df['year'] = df['DateTime'].dt.year
df['year'] = df['year'].astype('object')
```

```python
df = df[df['% Played Video'] != '---'] 
# from object to float
df['% Played Video'] = df['% Played Video'].astype(float)
```

### Get the number of every column instead of percentage.

```python
df['male'] = ((df['Participants (Course Content Accessed)'])*(df['% Male']/100)).astype(int)
df['female'] = ((df['Participants (Course Content Accessed)'])*(df['% Female']/100)).astype(int)
df['Audited'] = ((df['Participants (Course Content Accessed)'])*(df['% Audited']/100)).astype(int)
df['Posted_in_Forum'] = ((df['Participants (Course Content Accessed)'])*(df['% Posted in Forum']/100)).astype(int)
df['Grade_Higher_Than_Zero'] = ((df['Participants (Course Content Accessed)'])*(df['% Grade Higher Than Zero']/100)).astype(int)
df['Played Video'] = ((df['Participants (Course Content Accessed)'])*(df['% Played Video']/100)).astype(int)
df['B_degree_or_higher'] = (df['Participants (Course Content Accessed)'])*(df['% Bachelors Degree or Higher']/100)

df.rename(columns={'Participants (Course Content Accessed)': 'Participants'}, inplace=True)

df.drop(columns=['% Posted in Forum','% Audited','% Female','% Male','DateTime'],inplace=True)
df.head()
```
